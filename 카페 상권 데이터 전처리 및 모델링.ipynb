```python
a = s1df_c['시군구명'] # 항상 같으니깐 빼야함!

for i,element in enumerate(a.unique()):
    print('---------------------------------')
    print(element)
    corr = s1df_c[s1df_c['시군구명']==a.unique()[i]].corr(method='pearson')
    rank = corr['총_매출_금액'][corr['총_매출_금액'].rank()>=16].sort_values(ascending=False) # rank()가 15 이상인 항목을 상관관계 크기 순서대로 
    print(rank)
```

```python
corr
```

```python
corr.shape
```

```python
rounded_corr = corr.apply(lambda x: round(x ,3))
rounded_corr
```

```python
fig, ax = plt.subplots()
im = ax.imshow(rounded_corr, cmap='Greys')

# Color Bar
cbar = ax.figure.colorbar(im, ax=ax)

ax.set_xticks(np.arange(len(rounded_corr.columns)))
ax.set_yticks(np.arange(len(rounded_corr.index)))

ax.set_xticklabels(rounded_corr.columns)
ax.set_yticklabels(rounded_corr.columns)

for x in range(len(rounded_corr.columns)):
    for y in range(len(rounded_corr.index)):
        ax.text(y, x, rounded_corr.iloc[y, x], ha='center', va='center', color='g')

fig.tight_layout()   
plt.show() 
```

```python
corr10 = rounded_corr.nlargest(10, '총_매출_금액')
corr10 = corr10[list(corr10.index)]
corr10
```
