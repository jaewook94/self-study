```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.read_excel('C:/Users/asus/Desktop/8.27_상권_데이터_수정본.xlsx')
df = pd.DataFrame(df)
```

```python
# 2020년 데이터, 상권
s1df = df.copy()
```

```python
# 결측치 제거
s1df.isnull().sum()
```

```python
s1df = s1df.dropna()
```

```python
# 구별 상관관계 요소 상위 5개 구하기(매출 데이터 제외)
s1df_c = s1df.drop(['점포수', '폐업_점포_수', '프랜차이즈_점포_수', '주중_매출_금액',
                   '주말_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액',
                   '시간대_14~17_매출_금액', '시간대_17~21_매출_금액','연령대_10_매출_금액',
                   '연령대_20_매출_금액', '연령대_30_매출_금액', '연령대_40_매출_금액',
                   '연령대_50_매출_금액', '연령대_60_이상_매출_금액'], axis=1)
s1df_c.sample(5)
```

```python
for i in s1df_c:
    a = s1df_c['시군구명']
    corr = s1df_c[s1df_c['시군구명']==a.unique()[i-1]].corr(method='pearson')
    rank = corr['총_매출_금액'][corr['총_매출_금액'].rank()>=10]
```

```python
a=s1df_c.loc[:,'시군구명']
corr =s1df_c[s1df_c['시군구명']==a.unique()[0]].corr(method='pearson')
corr['총_매출_금액'][corr['총_매출_금액'].rank()>=10]
```

```python
# 구별로 묶기
# 합계 열 grouping
s1df_sum_group = s1df.groupby(['시군구명'])['점포수','프랜차이즈_점포_수','폐업_점포_수', '총_매출_금액', '주중_매출_금액',
                      '주말_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액',
                      '시간대_14~17_매출_금액', '시간대_17~21_매출_금액','연령대_10_매출_금액',
                      '연령대_20_매출_금액', '연령대_30_매출_금액', '연령대_40_매출_금액',
                      '연령대_50_매출_금액', '연령대_60_이상_매출_금액','아파트_단지_수',
                      '총 상주인구 수', '총 가구 수', '총_생활인구_수', '연령대_10_생활인구_수',
                      '연령대_20_생활인구_수', '연령대_30_생활인구_수', '연령대_40_생활인구_수',
                      '연령대_50_생활인구_수', '연령대_60_이상_생활인구_수', '주중_생활인구_수',
                      '주말_생활인구_수', '총_직장_인구_수', '연령대_10_직장_인구_수',
                      '연령대_20_직장_인구_수', '연령대_30_직장_인구_수', '연령대_40_직장_인구_수',
                      '연령대_50_직장_인구_수', '연령대_60_이상_직장_인구_수'].sum()
```

```python
# 평균 열 grouping
s1df_mean_group = s1df.groupby(['시군구명'])['개업_율', '폐업_률', '아파트_평균_면적',
                                        '아파트_평균_시가', '운영_영업_개월_평균',
                                        '폐업_영업_개월_평균'].mean()
```

```python
s2df = pd.concat([s1df_sum_group,s1df_mean_group], axis=1)
```

```python
# 총매출 제외 매출 칼럼 제거
s3df = s2df.drop(['주중_매출_금액', '주말_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액',
          '시간대_14~17_매출_금액', '시간대_17~21_매출_금액', '연령대_10_매출_금액',
          '연령대_20_매출_금액', '연령대_30_매출_금액', '연령대_40_매출_금액',
          '연령대_50_매출_금액', '연령대_60_이상_매출_금액'], axis=1)
```

```python
# 상관관계 분석
corr = s3df.corr(method='pearson')
```
